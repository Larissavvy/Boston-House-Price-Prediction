## The Boston Housing Dataset 

#### The Boston Housing Dataset is a derived from information collected by the U.S. Census Service concerning housing in the area of Boston MA. The following describes the dataset columns:

    CRIM - per capita crime rate by town
    ZN - proportion of residential land zoned for lots over 25,000 sq.ft.
    INDUS - proportion of non-retail business acres per town.
    CHAS - Charles River dummy variable (1 if tract bounds river; 0 otherwise)
    NOX - nitric oxides concentration (parts per 10 million)
    RM - average number of rooms per dwelling
    AGE - proportion of owner-occupied units built prior to 1940
    DIS - weighted distances to five Boston employment centres
    RAD - index of accessibility to radial highways
    TAX - full-value property-tax rate per (dollar)10,000
    PTRATIO - pupil-teacher ratio by town
    B - 1000(Bk - 0.63)^2 where Bk is the proportion of blacks by town
    LSTAT - % lower status of the population
    MEDV - Median value of owner-occupied homes in (dollar)1000's

### Models

```python
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn import tree 
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score , confusion_matrix , roc_auc_score , roc_curve
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor
from xgboost import XGBRegressor
from sklearn.ensemble import ExtraTreesRegressor
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error,mean_absolute_error
from sklearn.model_selection import cross_val_score
from sklearn import metrics
from sklearn.linear_model import Ridge
```

```python
X = df_scaled.drop(["MEDV"],axis=1)
y = df_scaled.MEDV
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=30)

print("X_train shape:", X_train.shape)
print("X_test shape:", X_test.shape)
print("y_train shape:", y_train.shape)
print("y_test shape:", y_test.shape)
```

```python
lr = LinearRegression()
rf = RandomForestRegressor()
dt = DecisionTreeRegressor()
xgb = XGBRegressor()
rr = Ridge()
```

```python
def model(mod,X_train,y_train,X_test,y_test):
    mod.fit(X_train,y_train)
    y_pred = mod.predict(X_test)
    
    #CrossValidation
    cv_model = cross_val_score(estimator = mod, X = X_train, y = y_train, cv = 15)
    
    #Train Data Predictions
    y_pred_train = mod.predict(X_train)
    r2_score_train = r2_score(y_train,y_pred_train)
    
    #Test Data Predictions
    y_pred_test = mod.predict(X_test)
    r2_score_test = r2_score(y_test,y_pred_test)
    
    #All
    rmse_model = np.sqrt((mean_squared_error(y_test, y_pred_test)))
    print("CV: ", cv_model.mean())
    print('R2_score [Accuracy of train]: ', r2_score_train*100)
    print('R2_score [Accuracy of test]: ', r2_score_test*100)
    print("RMSE: ", rmse_model)
    print('MAE: ', metrics.mean_absolute_error(y_test,y_pred_test))
    print('MSE: ', metrics.mean_squared_error(y_test, y_pred_test))
    
    sns.set_style("dark")
    #predictions for Train
    sns.scatterplot(y_train, y_pred_train,markers="*",color = "g")
    plt.xlabel("Prices")
    plt.ylabel("Predicted prices")
    plt.title("Prices vs Predicted prices for Train Data")
    plt.show()
    
    #predictions for Train
    sns.scatterplot(y_test, y_pred_test,markers="+",color = "r")
    plt.xlabel("Prices")
    plt.ylabel("Predicted prices")
    plt.title("Prices vs Predicted prices for Test Data")
    plt.show()
    
    
    # Checking Normality
    sns.distplot(y_train-y_pred_train,color = "purple")
    plt.title("Histogram of Residuals")
    plt.xlabel("Residuals")
    plt.ylabel("Frequency")
    plt.show()
    
```

```python
#1.Linear Regression
model(lr,X_train,y_train,X_test,y_test)
```

```python
# 2.Random Forest
model(rf,X_train,y_train,X_test,y_test)
```

```python
#3.Decision Tree
model(dt,X_train,y_train,X_test,y_test)
```

```python
#4.XGB
model(xgb,X_train,y_train,X_test,y_test)
```

```python
#5.ExtraTreeRegressor
reg= ExtraTreesRegressor()
reg.fit(X_train,y_train)
y_tree = reg.predict(X_test)
model(reg,X_train,y_train,X_test,y_test)
```

```python
#Random Forest Hyper Parameter:

n_estimators = [5,20,50,100]
max_features = ['auto', 'sqrt'] 
max_depth = [int(x) for x in np.linspace(10, 120, num = 12)] 
min_samples_split = [2, 6, 10] 
min_samples_leaf = [1, 3, 4] 

random_grid = {'n_estimators': n_estimators,

'max_features': max_features,

'max_depth': max_depth,

'min_samples_split': min_samples_split,

'min_samples_leaf': min_samples_leaf,}

from sklearn.model_selection import RandomizedSearchCV
rf_random = RandomizedSearchCV(estimator = rf,param_distributions = random_grid,
               n_iter = 200, cv = 5, verbose=2, random_state=35, n_jobs = -1)
```

```python
rf_random.fit(X_train, y_train)
```

```python
print ('Random grid: ', random_grid, '\n')
# print the best parameters
print ('Best Parameters: ', rf_random.best_params_, ' \n')
```
